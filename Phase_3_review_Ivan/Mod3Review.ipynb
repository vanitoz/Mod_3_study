```python
# What is a cost function? What are we trying to find?

# Cost function is over all cost of building a model.
# If the cost is very high - model does not represent data very well (model not good)
# If the cost very low - model represent data very well (good model)
```

```python
# How to use gd to find the lowest point? How does the gradient change as we get closer to the bottom?

# When using GD to cost curve  ùêΩ(ùëö,ùëè) , the technique should answer how much to move the  ùëö  variable and the  ùëè  variable to produce 
# the greatest decrease in cost, or RSS. In other words, when altering our regression line, we want to know how much of this change 
# should be derived from a move in the slope versus how much should be derived from a change in the y-intercept.

# As we getting closer to the bottom of the cost curve - gradient will change less and less closer to 0.

# learning rate regulates the size of the step we are taking towards the bottom of cost curve.
# High learning rate can cause missing the bottom of cost curve (jump over the minimum) and never finds minimum of the cost. 
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error
import seaborn as sns
%matplotlib inline
x = [1, 1, 2, 3, 4, 3, 4, 6, 4]
y = [2, 1, 0.5, 1, 3, 3, 2, 5, 4]
plt.figure(figsize = (8,8))
fig, ax = plt.subplots()

ax.scatter(x,y)
plt.show()
```

```python
beta_0 = 0
#beta_1 = [.25, .5, .75, .8, 1,]
beta_1 = np.arange(0.25,3,0.1)
fig, ax = plt.subplots(figsize=(10,6))
mses = []
for t in beta_1:
    line = beta_0 + (np.array(x)*t)
    mse = round(mean_squared_error(y, line),3)
    mses.append(mse)
    ax.plot(x, line, label=f'{mse} {t}')
ax.scatter(x,y)
plt.legend()
plt.show()
```

```python
# Plot the Cost Curve
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(beta_1, mses)
ax.set_title('Cost Curve')
ax.set_xlabel('beta 1')
ax.set_ylabel('MSE')
plt.show()
```

### Class Imbalance
<img src = 'imbalanced.png' wid = 300>

```python
### what problem would it cause? 

# Accuracy might not be appropriate metric for evaluation 
# Standard classifier algorithms like Decision Tree and Logistic Regression
# have a bias towards classes which have number of instances. They tend to only predict 
# the majority class data. The features of the minority class are treated as noise and are often ignored. Thus, there is a high probability of misclassification of the minority class as compared to the majority class.

```

```python
### How to remedy it? describe some detailed solution

## Resampling
### A common technique for dealing with highly unbalanced datasets is called resampling.
# It consists of removing samples from the majority class (under-sampling) and / or adding
# more examples from the minority class (over-sampling).


## Under-sampling: Tomek links

# Tomek links are pairs of very close instances, but of opposite classes.
# Removing the instances of the majority class of each pair increases the space between 
# the two classes, facilitating the classification process.


# Over-sampling: SMOTE
# SMOTE (Synthetic Minority Oversampling Technique) consists of synthesizing elements 
# for the minority class, based on those that already exist. It works randomly picking 
# a point from the minority class and computing the k-nearest neighbors for this point.
# The synthetic points are added between the chosen point and its neighbors.
```

## Part III. Decision Trees
- Build trees with the sklearn machine learning framework


How do they make predictions - voting on target class

use entropy or gini to determine split on the most powerful feature, make another split after on the most greedy features divide on attribute at a time in a greedy manner to create homogenous groups (the word pure freaks everyone out) make splits - determine most predict attribute (entropy/gini), then continue with the next most predictive attribute until hyperparams been reach
make predictions - running through the decision tree

Decision Trees How do decision trees work? Divides data based on one attribute/feature at a time in a greedy manner to create more homogenous groups. How do they make splits? Determine the attribute with the most predictive power (greedy model), ie that produces the lowest cost function (gini or entropy) based on that split alone. Then continue with the next most predictive attribute, etc, until specified hyperparameters have been reached or homogeneity has been achieved. How do DTs make predictions? Run testing data thru the decision tree that was created via the training data

```python
# import the dataset and set up predictors and target
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# define x and y 
y = titanic['survived']
X = titanic[['pclass','age', 'sibsp', 'parch', 'fare', 'adult_male']]
```

```python
# fill the age columns missing value with mean 
X = X.fillna(round(X.age.mean()))
```

```python
## check the info and see if it's ready to be modeled
X.info()
```

```python
# Train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# fit the tree
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
# test the tree using .predict method
y_pred_test = clf.predict(X_test)
```

```python
# generate prediction and output metric (use accuracy)
from sklearn.metrics import accuracy_score

print('Accuracy: ', accuracy_score(y_test, y_pred_test))
```
